```python
drop_columns = []
for column in data.columns:
    corrcoef = np.corrcoef(data['Class'],data[column])
    if np.abs(corrcoef[0,1]) < 0.02:
        drop_columns.append(column)
print(drop_columns)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler

X = data.drop(columns='Class')
y = data['Class']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=123, stratify= y)
X_train_drop = X_train.drop(columns=drop_columns)
X_test_drop = X_test.drop(columns=drop_columns)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.metrics import make_scorer
```

```python
log = LogisticRegression()
Knn = KNeighborsClassifier(n_neighbors=5)
```

```python
Knn.fit(X_train_scaled,y_train)
Knn_pred_scaled = Knn.predict(X_test_scaled)
Knn.fit(X_train_drop,y_train)
Knn_pred_drop = Knn.predict(X_test_drop.values)
log.fit(X_train_scaled,y_train)
log_pred_scaled = log.predict(X_test_scaled)
log.fit(X_train_drop,y_train)
log_pred_drop = log.predict(X_test_drop)
```

```python
result_report = pd.DataFrame()
result_report['model'] = ['knn_scaled','knn_drop','log_scaled','log_drop']
m = [Knn_pred_scaled,Knn_pred_drop,log_pred_scaled,log_pred_drop]
a_score = []
p_score = []
r_score = []
cm = []
for x in m:
    a = accuracy_score(y_test,x)
    a_score.append(a)
    p = precision_score(y_test,x)
    p_score.append(p)
    r = recall_score(y_test,x)
    r_score.append(r)
    cmatrix = confusion_matrix(y_test,x)
    cm.append(cmatrix)
result_report['accuracy_score'] = a_score
result_report['precision_score'] = p_score
result_report['recall_score'] = r_score
result_report['confusion_matrix'] = cm
result_report.set_index('model',inplace=True)
print(result_report)
```
